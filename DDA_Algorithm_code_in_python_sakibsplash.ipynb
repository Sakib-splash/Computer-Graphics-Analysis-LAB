```python

#code by sakib-11
import matplotlib.pyplot as plt
#function for finding the intermediary value
def DDA(x0,y0,x1,y1):

    #difference of x
    dx=abs(x0-x1)
    
    #difference of x
    dy=abs(y0-y1)
    
    #finding max steps for the points
    steps=max(dx,dy)
    
    xsteps= dx/steps
    ysteps= dy/steps
    x= float(x0)
    y= float(y0)
    #creating the list to store intermediary  x,y values
    x_points=[]
    y_points=[]
    for i in range(steps):
        #inserting in list
        x_points.append(x)
        y_points.append(y)
        #increment those x,y
        x=x+xsteps
        y=y+ysteps

    plt.plot(x_points,y_points,marker="o", markersize=1, markerfacecolor="green")
    plt.show()
#main code
x0,y0=20,20
x1,y1=60,50
DDA(x0,y0,x1,y1)
```
