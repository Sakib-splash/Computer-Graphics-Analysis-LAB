```python
import matplotlib.pyplot as plt
#code by sakib11
def midpointLine(x1,y1,x2,y2):
    dx=x2-x1
    dy=y2-y1
    #calculating d for finding the E or NE
    d=dy-(dx/2)
    x=x1
    y=y1
    #creating list for the midpoint values
    x_points=[]
    y_points=[]
    x_points.append(x)
    y_points.append(y)
    print(x,",",y,"\n")
    while(x<x2):
        x=x+1
        if(d<0):
            #select East point
            d=d+dy
        else:
            #select NE point
            d=d+(dy-dx)
            y=y+1
        print(x,",",y,"\n")
        #inserting the new x,y values in the list
        x_points.append(x)
        y_points.append(y)
    #plotting that are visualized in graph
    plt.plot(x_points,y_points,marker="o", markersize=1, markerfacecolor="green")
    plt.show()
#inserting the two points that are given
x1=3
y1=3
x2=9
y2=6
#calling the midpoint function
midpointLine(x1,y1,x2,y2)
```
